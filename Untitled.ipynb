```python
import numpy as np
from PIL import Image
from env import VirtualLD19

# Load your map
img = np.array(Image.open('map.png').convert('L'))
print(f"Map size: {img.shape}")
print(f"Pixel value range: {img.min()} to {img.max()}")

# Convert to walls (1) and free space (0)
# We treat anything darker than mid-grey as a wall
grid = (img < 240).astype(np.uint8)
print(f"Walls: {grid.sum()} pixels, Free: {(grid==0).sum()} pixels")

# Create the lidar
lidar = VirtualLD19(grid, resolution=0.05)

# Try a scan from somewhere in your map
# Pick a position - we need to find a free spot first
free_pixels = np.argwhere(grid == 0)
py, px = free_pixels[len(free_pixels) // 2]  # pick a middle one
x = px * 0.05
y = py * 0.05
print(f"\nScanning from world position ({x:.2f}m, {y:.2f}m)")
print(f"  (which is pixel ({px}, {py}) on the map)")

# Run a scan
scan = lidar.scan(x, y, theta=0.0)

print(f"\nScan results:")
print(f"  Number of rays: {len(scan)}")
print(f"  Min distance: {scan.min():.2f}m")
print(f"  Max distance: {scan.max():.2f}m")
print(f"  Mean distance: {scan.mean():.2f}m")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy import ndimage
from env import VirtualLD19

# Load map
img = np.array(Image.open('map.png').convert('L'))
grid = (img < 240).astype(np.uint8)
resolution = 0.05

# Create lidar
lidar = VirtualLD19(grid, resolution)

# Find a spot far from any wall using distance transform
# This finds the pixel that's deepest inside the corridor
dist_to_wall = ndimage.distance_transform_edt(grid == 0)
py, px = np.unravel_index(np.argmax(dist_to_wall), dist_to_wall.shape)
x, y = px * resolution, py * resolution
theta = 0.0

print(f"Spawning car at world ({x:.2f}m, {y:.2f}m)")
print(f"Distance to nearest wall: {dist_to_wall[py, px] * resolution:.2f}m")

# Scan
scan = lidar.scan(x, y, theta)

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(img, cmap='gray',
          extent=[0, grid.shape[1] * resolution,
                  grid.shape[0] * resolution, 0])

# Draw lidar rays
for i, ray_angle in enumerate(lidar.ray_angles):
    world_angle = theta + ray_angle
    end_x = x + scan[i] * np.cos(world_angle)
    end_y = y + scan[i] * np.sin(world_angle)
    ax.plot([x, end_x], [y, end_y], 'b-', alpha=0.3, linewidth=0.5)

# Draw car
ax.plot(x, y, 'ro', markersize=10)
ax.plot([x, x + 0.5*np.cos(theta)], [y, y + 0.5*np.sin(theta)], 'r-', linewidth=2)

ax.set_title(f'Virtual LD19 scan from open area (450 rays, max range {lidar.max_range}m)')
ax.set_xlabel('X (meters)')
ax.set_ylabel('Y (meters)')
plt.savefig('scan_test2.png', dpi=80)
plt.show()
```

```python
import numpy as np
from env import LidarCarEnv

# Create environment
env = LidarCarEnv('map.png')

print("=" * 50)
print("Environment created successfully")
print(f"Observation space: {env.observation_space.shape}")
print(f"Action space: {env.action_space.shape}")
print("=" * 50)

# Reset
obs, info = env.reset()
print(f"\nFirst observation:")
print(f"  Shape: {obs.shape}")
print(f"  Range: [{obs.min():.3f}, {obs.max():.3f}]")
print(f"  First 5 values (lidar): {obs[:5]}")
print(f"  Velocity (index 50): {obs[50]:.3f}")
print(f"  Prev action (last 2): {obs[51:]}")

# Drive forward for a few steps
print(f"\nDriving forward (action=[0, 0.8]):")
for i in range(10):
    action = np.array([0.0, 0.8], dtype=np.float32)  # straight + accelerate
    obs, reward, term, trunc, info = env.step(action)
    print(f"  Step {i+1}: pos=({info['x']:.2f}, {info['y']:.2f}), "
          f"v={info['v']:.2f}m/s, reward={reward:+.3f}, "
          f"collided={info['collided']}")
    if term:
        print("  CRASHED!")
        break

# Try a random policy episode
print(f"\nRandom action episode:")
obs, info = env.reset()
total_reward = 0
for step in range(100):
    action = env.action_space.sample()  # random action
    obs, reward, term, trunc, info = env.step(action)
    total_reward += reward
    if term or trunc:
        break
print(f"  Survived {step+1} steps, total reward: {total_reward:+.2f}")
print(f"  Ended because: {'collision' if term else 'timeout'}")
```

```python
"""
Trains a PPO agent on the LidarCarEnv.
Run: python train.py
"""
import os
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack, VecMonitor
from stable_baselines3.common.callbacks import EvalCallback, CheckpointCallback
from env import LidarCarEnv


def make_env():
    """Factory function that creates one environment instance."""
    return LidarCarEnv('map.png')


# --- Set up the training environment ---
# DummyVecEnv: wraps our env in SB3's vectorized format (required)
# VecMonitor: tracks episode rewards and lengths
# VecFrameStack: stacks the last 4 observations so the policy can sense motion
train_env = DummyVecEnv([make_env])
train_env = VecMonitor(train_env)
train_env = VecFrameStack(train_env, n_stack=4)

# --- Set up an eval environment (separate copy) ---
eval_env = DummyVecEnv([make_env])
eval_env = VecMonitor(eval_env)
eval_env = VecFrameStack(eval_env, n_stack=4)

# --- Callbacks ---
# Save the best model whenever eval reward improves
eval_callback = EvalCallback(
    eval_env,
    best_model_save_path='./best_model/',
    log_path='./logs/eval/',
    eval_freq=10_000,        # evaluate every 10k training steps
    n_eval_episodes=5,        # average over 5 episodes
    deterministic=True,
    render=False,
)

# Also save periodic checkpoints in case training crashes
checkpoint_callback = CheckpointCallback(
    save_freq=50_000,
    save_path='./checkpoints/',
    name_prefix='ppo_lidar_car',
)

# --- Create the PPO model ---
model = PPO(
    'MlpPolicy',
    train_env,
    verbose=1,
    tensorboard_log='./logs/',
    
    # Key hyperparameters (these are SB3 defaults, listed for clarity)
    learning_rate=3e-4,       # how big each gradient step is
    n_steps=2048,             # collect 2048 steps before updating
    batch_size=64,            # minibatch size for SGD
    n_epochs=10,              # passes over collected data per update
    gamma=0.99,               # discount factor (cares ~100 steps ahead)
    gae_lambda=0.95,          # GAE smoothing parameter
    clip_range=0.2,           # PPO's safety mechanism
    ent_coef=0.01,            # exploration bonus
    
    device='cpu',            # uses GPU if available, else CPU
)

print("=" * 60)
print(f"Training device: {model.device}")
print(f"Total training steps: 1,000,000")
print(f"Estimated time: 1-3 hours")
print("=" * 60)

# --- Train ---
model.learn(
    total_timesteps=1_000_000,
    callback=[eval_callback, checkpoint_callback],
    progress_bar=True,
)

# --- Save the final model ---
model.save('final_model')
print("\nTraining complete! Saved as final_model.zip")
print("Best model saved in best_model/best_model.zip")
```
